```python
# y^2-x-2y+1
# x'^2+2\sqrt{3}x'y'+3y'^2-x'(2\sqrt{3}+4)-y'(4\sqrt{3}-2)+4

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['lines.color'] = 'k'
mpl.rcParams['axes.prop_cycle'] = mpl.cycler('color', ['k'])

x = np.linspace(-9, 9, 400)
y = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x, y)

def axes():
    plt.axhline(0, alpha=.1)
    plt.axvline(0, alpha=.1)

a, b, c, d, e, f =  0, 0, 1, -1, -2, 1
a_r, b_r, c_r, d_r, e_r, f_r =  1, 2*np.sqrt(3), 3, -1*(2*np.sqrt(3)+4), -1*(4*np.sqrt(3)-2), 4,
# assert b**2 - 4*a*c < 0
axes()
plt.contour(x, y,(a*x**2 + b*x*y + c*y**2 + d*x + e*y + f), [0], colors='#aa0000')
plt.contour(x, y,(a_r*x**2 + b_r*x*y + c_r*y**2 + d_r*x + e_r*y + f_r), [0], colors='#0000dd')
plt.show()
```
